```python
import cv2

import numpy as np
img = cv2.imread('C:\\Users\\arask\\source\\repos\\workspace\\cv_lab03\\robot_010.png', cv2.IMREAD_GRAYSCALE)
kernel = np.ones((9,9),np.float32)/81
img_conv = cv2.filter2D(img,-1,kernel)

cv2.imshow('Original', img)
cv2.imshow('Convolution', img_conv)
```

```python
sobel_x = np.array([[1, 0, -1],
[2, 0, -2],
[1, 0, -1]])
sobel_y = np.array([[1, 2, 1],
[0, 0, 0],
[-1, -2, -1]])
import numpy as np
img = cv2.imread('C:\\Users\\arask\\source\\repos\\workspace\\cv_lab03\\robot_010.png', cv2.IMREAD_GRAYSCALE)
kernel = sobel_x

kernel2 = sobel_y
img_conv1 = cv2.filter2D(img,-1,kernel)
img_conv2 = cv2.filter2D(img,-1,kernel2)

#cv2.flip(kernel, -1, kernel)#flip kernel do this when you are using asymmetric kernel
#cv2.flip(kernel2, -1, kernel2)
flipped_kernel = cv2.flip(kernel, -1)
flipped_kernel2 = cv2.flip(kernel2, -1)
img_conv_proper1 = cv2.filter2D(img,-1,flipped_kernel)
img_conv_proper2 = cv2.filter2D(img,-1,flipped_kernel2)
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 5, figsize=(20, 4))

axes[0].imshow(img, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(img_conv1, cmap='gray')
axes[1].set_title('Convolution Image sob x')
axes[1].axis('off')
axes[2].imshow(img_conv2, cmap='gray')
axes[2].set_title('Conv sob y')
axes[2].axis('off')
axes[3].imshow(img_conv_proper1, cmap='gray')
axes[3].set_title('Conv sob x proper')
axes[3].axis('off')
axes[4].imshow(img_conv_proper2, cmap='gray')
axes[4].set_title('Conv sob y proper')
axes[4].axis('off')
plt.tight_layout()
plt.show()
```

```python
newimg = img_conv_proper1 + img_conv_proper2
fig, axes = plt.subplots(1, 2, figsize=(20, 4))
axes[0].imshow(img_conv_proper2, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(newimg, cmap='gray')
axes[1].set_title('Convolution Image sob x')
axes[1].axis('off')
plt.show()
```

```python
img = cv2.imread('C:\\Users\\arask\\source\\repos\\workspace\\cv_lab03\\robot_010.png', cv2.IMREAD_COLOR)
# Now convert the image to RGB rather than BGR, for convenience when plotting it
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Create a new image the same size as the original
new_img = np.zeros(img.shape, np.uint8)
# Iterate over the pixels in the image
# (i.e. the elements in the underlying array)
# and mark red pixels in the new image

# Remember OpenCV 's colour channels are B,G,R so the red channel is index 2
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
# Channel 0 is red in RGB format
        if img[i,j,0] >= 250:
            new_img[i,j,0] = 255
# Code to show images is omitted for brevity. You know how to do that.
fig, axes = plt.subplots(1,2, figsize=(20, 4))
axes[0].imshow(img)
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(new_img)
axes[1].set_title('Red Pixels')
axes[1].axis('off')
plt.show()
```

```python
import cv2
def sliderCallback(x):
    print(x)
# Create a window, and give it a name
cv2.namedWindow('Tracker sample',cv2.WINDOW_AUTOSIZE)
# Create a trackbar
# - give it a name ('Red')
# - attach it to the named window ('Tracker sample')
# - set its minimum and maximum values (0 and 255)
# - give it the name of a function to call when it changes (sliderCallback)
cv2.createTrackbar('Red', 'Tracker sample', 0, 255, sliderCallback)
while(1):
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
minred=0
has_changed=True
def sliderCallback(x):
# need to declare variables used callbacks and main program as global   
    global minred
    global has_changed
    minred=x
    has_changed=True

print(x)
img=cv2.imread('data/robot_010_sm.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
new_img=np.zeros(img.shape, np.uint8)
cv2.namedWindow('Tracker sample',cv2.WINDOW_AUTOSIZE)
cv2.createTrackbar('Red', 'Tracker sample', 0, 255, sliderCallback)
while(1):
# Only process the image if the slider value has changed
if has_changed:
has_changed=False
for i in range(img.shape[0]):
for j in range(img.shape[1]):
if img[i,j,0] >= minred:
new_img[i,j,2] = 255
else:
new_img[i,j,2] = 0
cv2.imshow('Tracker sample', new_img)
if cv2.waitKey(1) & 0xFF == ord('q'):
break
cv2.destroyAllWindows()
```
